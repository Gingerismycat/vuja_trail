```python
import random
import time
print('Hello! You, a farmer, decide that you want to embark on the Vuja Trail to make it to Texas. Beware this trail is full of danger, so you brought a carpenter, hunter, and missionary!')
time.sleep(.5)
print('''Day 1: 
+100 pounds of food
+2 Oxen
+1 Bible (Morale Booster)
+1 Wagon
+1 Gun
+10 Ammo
+100 Dollars
+1 Bible
''')
#Character Conditions
class playercondition():
  def __init__(self):
    self.stamina = 100
    self.morale = 100
    self.health = 100
    self.hunger = 100
class carpentercondition():
  def __init__(self):
    self.stamina = 100
    self.morale = 100
    self.health = 100
    self.hunger = 100
class huntercondition():
  def __init__(self):
    self.stamina = 100
    self.morale = 100
    self.health = 100
    self.hunger = 100
class missionarycondition():
  def __init__(self):
    self.stamina = 100
    self.morale = 100
    self.health = 100
    self.hunger = 100

#Game
class RunGame():
  def __init__(self):
    self.day = 1
    self.time = 1
    self.steps = 0
    self.wagoncondition = 100
    self.player_condition = playercondition()
    self.carpenter_condition = carpentercondition()
    self.hunter_condition = huntercondition()
    self.missionary_condition = missionarycondition()
    self.inventory = {
      "food":100,
      "oxen":2,
      "gun" :1,
      "ammo":10,
      "money":100,
      "entertainmentitem": 1
    }
    while True:
      user_input = input("What would you like to do?\n1) Walk\n2) Rest\n3) Check Status\n")
      user_input = int(user_input)
      time.sleep(.5)
      if self.steps < 100000:
        if user_input == 1:
          print("You decided to keep going forward.")
          self.steps = self.steps + 100
          self.time = self.time + random.randint(5,10)
          self.inventory["food"] = self.inventory["food"] - 10
          time.sleep(.5)
          print(f'{self.steps} steps taken out of 100000')
          time.sleep(.5)
          self.walkevent()
        elif user_input == 2:
          print("You decided to rest for now")
          self.time = self.time + 10
        elif user_input == 3:
          user_input = input('''What would you like to check?\n1) Inventory\n2) Party Condition\n3) Wagon Condition\n''')
          user_input = int(user_input)
          if user_input == 1:
            print(self.inventory)
          elif user_input == 2:
            print(f"You: Stamina - {self.player_condition.stamina}, Morale - {self.player_condition.morale}, Health - {self.player_condition.health}, Hunger - {self.player_condition.hunger}")
            print(f"Carpenter: Stamina - {self.carpenter_condition.stamina}, Morale - {self.carpenter_condition.morale}, Health - {self.carpenter_condition.health}, Hunger - {self.carpenter_condition.hunger}")
            print(f"Hunter: Stamina - {self.hunter_condition.stamina}, Morale - {self.hunter_condition.morale}, Health - {self.hunter_condition.health}, Hunger - {self.hunter_condition.hunger}")
            print(f"Missionary: Stamina - {self.missionary_condition.stamina}, Morale - {self.missionary_condition.morale}, Health - {self.missionary_condition.health}, Hunger - {self.missionary_condition.hunger}")
          elif user_input == 3:
            print(f"The wagon is at {self.wagoncondition}%")
          else:
            print("Input a number that is avaliable!")
        else:
          print("Input a number that is avaliable!")
      else:
        print("YOU MADE IT!!!")
  def walkevent(self):
    event = random.randint(1,100)
    if event <= 10:
      amountofberries = random.randint(1,20)
      self.inventory["food"] = self.inventory["food"] + amountofberries
      print(f"You found {amountofberries} berries!")
    if event >= 11 and event <= 20:
      print("You encountered a river!")
      if self.rivercrossing() is False:
        self.inventory["food"] = int(self.inventory["food"]/2)
        print('The wagon got damaged and you lost half of your food!')
        self.wagoncondition = self.wagoncondition - random.randint(10,20)
      else:
        print('You successfully crossed the river!')
        self.wagoncondition = self.wagoncondition - 5
  def rivercrossing(self):
    if self.wagoncondition == 100:
      return True
    elif self.wagoncondition > 100:
      event = random.randint(1,100)
      if event > self.wagoncondition:
        return False
      else:
        return True
rungame = RunGame()
```
